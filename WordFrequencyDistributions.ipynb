Word frequency distributions are characterized by a small number of very high frequency words (with ranks close to 1) and large numbers of low frequency words. The lowest frequency words, the words with frequency 1, show up as the rightmost horizontal line in the rank-frequency plot (These words are known as hapax legomena (Greek for words read once in a text).

**Zipf law**
* The intuition is that frequency decreases very rapidly with rank.
* The collection frequency is inversely proportional to rank in the freq table.
* There is a power law between frequency and rank.

$$
\begin{align*}
\displaystyle f_r & \propto r^{-s} \\
\displaystyle f_r & \propto \frac{1}{r^s}
\end{align*}
$$ with $s>0$

* Equation above is one of the simplest ways of formalizing such a rapid decrease and it has been found to be a reasonably good model.
* s is called the *power law exponent*.
* if s = 1 and if the most frequent term occurs f_1 times, then the second most frequent term has half as many occurrences, the third most frequent term a third as many occurrences, and so on.

**log-log planes**
* The most frequently used graphical method for identification of a power law probability is log-log plotting.
* A straight line on a loglog plot is necessary (but insufficient) evidence for power-laws.
* The slope of the straight line corresponds to the power law exponent.")

```python
plt.loglog(ranks, sorted_freqs, drawstyle="steps")
plt.xlabel("log rank")
plt.ylabel("log frequency")
plt.show()
```

We may change the proportional relation to an equality by including a constant $b$

$$ f_r = br^{-s} $$ with $b > 0$

**Probability distribution**

Let random variable $X$ represent the rank of a word randomly selected from the sorted list of $V$ word types:
$$X = r \text{ where } r \in \{1, 2, \cdots, V\}$$
Then probabiliy mass function $p_X$ that describes the distribution of $X$ can be estimated with relative frequecies as:
\begin{align*}
p_X(x=r)&=\frac{f_r}{\sum_{i=1}^V f_i}\\
    &= \frac{f_r}{N} \\
    &= \frac{br^{-s}}{N} \\
    &= b'r^{-s}
\end{align*}

so there is also a power law between probability and rank.

**discrete Zeta distribution**

* Zipfs power law brings us to a new kind of discrete probability distribution, called the **Zeta** distribution.
* Zipf showed that the distribution of words in English follows a Zeta distribution.

* parameter: $s > 1$ 
* discrete random variable $X$ with PMF: 

    $$p_{X}(r)={r^{{-s}} \over \zeta (s)}$$

where $\zeta(s)$ is the Riemann zeta function:

$$  \zeta(s) = \frac{1}{1} + 
    \left(\frac{1}{2}\right)^{s} +
    \left(\frac{1}{3}\right)^{s} +
    \ldots +
    \left(\frac{1}{n}\right)^{s} + \ldots\;$$

```python
pr = dict()
for word, freq in word_freqs.items():
    pr[word] = freq/N
```

```python
pr
```

```python
pr["alice"]
```

```python
# sorted probabilities
sorted_pr = list(sorted(pr.values(), reverse=True))
```

```python
sorted_pr
```

```python
plt.plot(ranks, sorted_pr)
plt.xlabel("rank")
plt.ylabel("probability")
plt.show()
```

```python
plt.loglog(ranks, sorted_pr, drawstyle="steps")
plt.xlabel("log rank")
plt.ylabel("log probability")
plt.show()
```

As we inferred earlier, 

$$p_X(x=r) = b'r^{-s}$$

In other words rescaling frequencies to probabilities amounts to a change that can be absorbed into the b parameter of the power law.

```python
def pmf(r):
    return sorted_pr[r]

def cdf(r):
    return sum(sorted_pr[x] for x in range(r+1))

print(f'probability of seeing the most frequent word is {pmf(0)}')
print(f'probability of seeing the word with rank 10 is {pmf(9)}')
print(f'probability of seeing one of the top 10 words is {cdf(9)}')
```

The first 10 words constitute $\frac{1}{4}$-th of the language!
